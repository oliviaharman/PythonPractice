```python
rings = []
for smiles in comp_smiles: 
    m = rdkit.Chem.MolFromSmiles(smiles) # Must create a function for the code to act on ()
    ri = m.GetRingInfo() # Refer to rdkit tutortial website for the specific commands
    n_rings = ri.NumRings()
    rings.append(n_rings)
print(max(rings))
print(min(rings))
# use the rdkit website and stackoverflow to debug code.
# Remember to use functions correctly, they require () at the end of the command to run 
# Created a "for" loop within the list since we want to do something to every item in the list
# append adds a new item to the end of the list
```

```python
rings # Rings was the variable created from the last command
# Run the variable to store it and make sure the data looks correct
```

```python
plt.hist(rings, bins=5, density=True, facecolor='m', alpha=0.4)

# Rings is the variable created using the rdkit package.
# Density: if true, the data is normalized.
# Facecolor allows you to set the color of the bars on the histogram.
# Alpha is how darkly shaded in the bars on the histogram will be. Closer to 1.0 is more opaque.
# Different color options: 'r'=red, 'c'=cyan, 'm'=magenta, 'k'=black, 'g'=green 
# Matplotlib.org has a list of all the colors

plt.xlabel('Number of Rings in Molecule')

# Creates the xlabel on the histogram

#plt.xticks([0,1,2,3,4])

# Allows you to change the values along the x-axis.

plt.ylabel('BDE') # not sure what to label this axis (???)
plt.title('Histogram of Ring Numbers in Each Molecule')

# Creates the title of the histogram.

plt.rcParams["figure.figsize"] = (10,4)

# Command above allows you to change the width and height of the plot created.
# Can come in very handy when creating figures for poster.

plt.show

#When you run this code the histogram is generated, this command should always be at the end.
```
